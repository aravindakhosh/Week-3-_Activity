```python
# importing pandas to use as pd ,matplotlib.pyplot as plt and seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# read the the given data set  from file location
data = pd.read_csv(r"C:\Users\hp\OneDrive\Desktop\corse detail\diabetes_model.csv")
data.columns
```

```python
# Display the Glucose level and the Age for the samples having the index ranging from 45 to 60.
required_data = data[['Glucose','Age']]
```

```python
# to find How many patients in the data are having the blood pressure level above 120
High_BP = data.loc[data['BloodPressure'] > 120]
```

```python
# To Select the rows the age is between 24 and 29 (inclusive)
age_group = data[data['Age'].isin([24, 29])] 
age_group
```

```python
#Find the minimum and maximum value of insulin level in the data.
Minimum_insulin =data["Insulin"].min()
Maximum_insulin =data["Insulin"].max()
print("Minimum value of insulin level:",Minimum_insulin, "\nMaximum value of insulin level:", Maximum_insulin)
```

```python
#Create a histogram of "Age".
data.hist(column='Age',bins =20 )
plt.xlabel("Age")
plt.ylabel("Count")
plt.title(" Age")
```

```python
# to Create a scatter plot of "BMI" vs "Glucose".
x = data.BMI
y = data.Glucose
plt.xlabel("BMI")
plt.ylabel("Glucose")
plt.title(" BMI X Glucose")
plt.scatter(x,y, s=15, c='b')
plt.show
```

```python
# To Replace the min value of "BMI" with mean value of BMI
min_bmi=data["BMI"].min()
mean_BMI = data["BMI"].mean()
print("Minimum value of BMI:",min_bmi, "\nMaximum value of insulin level:", mean_BMI)
```

```python
data.replace(0,data.mean(axis=0),inplace=True)
```

```python
# To Draw a grouped boxplot with "Outcome" on x-axis and "Glucose" on the y-axis such that we see distributions of Glucose separately for each outcome.

plt.rcParams["figure.figsize"] = [20.00, 15.50]
plt.rcParams["figure.autolayout"] = True


sns.boxplot(x=data['Outcome'], y=data['Glucose'], hue=data['Age'])
plt.show()
```
